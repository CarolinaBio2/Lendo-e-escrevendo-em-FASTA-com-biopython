```python
import Bio
from Bio import SeqIO  #para fazer o parse e escrever
from Bio.Seq import Seq #para transformar uma sequência em um objeto do tipo seq (sequência de DNA)
from Bio.SeqRecord import SeqRecord  #para criar a ficha do genoma
```

```python
#LEITURA E FILTRAGEM DE ARQUIVOS EM FORMATO FASTA
myseq = Seq("ATCCGTAAACHTTAGGGCTAAATGGCGTAGGATCGGCATGCAAGGTCTTCAT")
gene = SeqRecord(myseq, id="gene_id_1", description="Descrição do gene 1")  #definindo uma variável chamada gene com a sequência criada e atribuindo uma descrição a ela
SeqIO.write(gene, "output.fna", "fasta")  #salvando um arquivo em seqIO.write no formato fasta e com o nome de output.fna, com a propriedade de escrita no arquivo
#um arquivo de proteínas não possui qualificadores de localização como "pseudo=true", "gbkey", "cds"
```

```python
for seq_record in SeqIO.parse("cds_from_genomic.fna", "fasta"):
  print(seq_record.description)
  #neste caso queremos acessar a descrição e pegar somente os qualificadores que são "pseudo=true"
```

```python
#quando abrimos um arquivo no modo escrita(write) que já tem alguma coisa dentro, o código vai abrir o arquivo, apagar tudo que tem nele, e escrever oque você quer, então a cada sequência, ele vai abrir, apagar e escrever sucessivamente, até o final do arquivo
#exemplo:
for seq_record in SeqIO.parse("cds_from_genomic.fna", "fasta"):
  if 'pseudo=true' in seq_record.description:
    myseq = seq_record.seq
    gene = SeqRecord(myseq, id=seq_record.id, description=seq_record.description)
    SeqIO.write(gene, "output_PSEUDO.fna", "fasta") #aqui foi salvo um arquivo (output_PSEUDO) que contém a descrição do último gene, ou seja, durante a leitura,o código foi abrindo, apagando e escrevendo tudo oque havia no arquivo, até chegar no último gene
```

```python
#para criar o arquivo a apenas uma vez e passar apenas a variável do arquivo aberto:
out = open("output_PSEUDO.fna", 'w')
for seq_record in SeqIO.parse ("cds_from_genomic.fna", "fasta"):
  if 'pseudo=true'in seq_record.description:
    myseq = seq_record.seq
    gene = SeqRecord(myseq, id=seq_record.id,description=seq_record.description)
    SeqIO.write(gene, out, "fasta")  #out é a variável criada para abrir o arquivo output_pseudo no formato fasta e modo escrita
out.close()  #fechar arquivo
#aqui criamos uma pasta com as sequências que possuem o qualificador "pseudo=true" em sua descrição; isso é uma espécie de filtragem
```

```python
#colocando várias sequencias de vários arquivos com o identificador pseudo=true em um único arquivo
import glob #oque isso mesmo?
for file in glob.glob("*.fna"): #o simbolo "*" significa "para todo arquivo de", que aqui nesse caso é "para todo arquivo de fasta"
  out = open(file.replace(".fna", "_OUT.fasta"), 'w') #substituindo o modo .fna pelo out.fasta; variável out dentro do laço for para criar um out put para cada arquivo
  for seq_record in SeqIO.parse("cds_from_genomic.fna", "fasta"):
    if 'pseudo=true' in seq_record.description:
      myseq = seq_record.seq
      gene = SeqRecord(myseq, id=seq_record.id, description=seq_record.description) #definindo a variável com a sequencia, id e descrição dos genes
      SeqIO.write(gene, out, "fasta") #abrir as duasvariáveis gene e out com o modo escrita e no formato fasta
#são criadas duas pastas cada uma correspondente ao seu formato
```
